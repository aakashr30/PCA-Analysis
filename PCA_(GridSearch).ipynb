```python
#load libraries
from sklearn import decomposition,datasets
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
#load the iris flower data
dataset = datasets.load_iris()
```

```python
x = dataset.data
y = dataset.target
```

```python
print(x)
print(y)
```

```python
#create an scaler object
sc = StandardScaler()
```

```python
#create a pca object
pca = decomposition.PCA()
```

```python
#create a decision tree Regression object with an L2 penalty
decisiontree = tree.DecisionTreeClassifier()
```

```python
#create a pipeline of three steps.first,standard the data,
#second, transform the data with pca
#third, train a Decision Tree Classifier on the data
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[('sc',sc),
                       ('pca',pca),
                       ('decisiontree',decisiontree)])
```

```python
#Create Parameter Space
#create the kist of a sequence of integers from 1 to 30 (the number of features of j)
n_components = list(range(1,x.shape[1]+1,1))
```

```python
#create lists of parameter for Decision Tree Classifier
criterion = ['gini','entropy']
max_depth = [4,6,8,12]
```

```python
#create  a dictionary of all the parameter
#note has you can acessthe parameters of steps of pipeline by using'_'
parameters = dict(pca__n_component=n_components,
                  decisiontree__criterion=criterion,
                  decisiontree__max_depth=max_depth)
```

```python
#conduct
clf = GridSearchCV(pipe, parameters)
```

```python
#fit thegrid search
clf.fit(x,y)
```

```python
#view thw best parameter
print('Best Criteron:',clf.best_estimator_.get_params()['decisiontree__criterion'])
```
